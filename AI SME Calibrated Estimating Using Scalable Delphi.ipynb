# Workshop Step 1: Quick Start

**Note (Requirements):**
- Recommended: Conda or Mamba installed (for `environment.yml`).
- Fallback: Any Python 3.11+ environment with `pip` (uses `requirements.txt`).
- Required at runtime: one API key for OpenAI (`OPENAI_API_KEY`).
- Optional local alternative: Ollama running on your machine (no API key needed).

1. Run Cell 2 to auto-install dependencies into the active kernel.
2. Run Cell 3 to verify imports and API key variables.

4. Run Cell 5 and continue with the notebook.

```python
# Workshop Step 2: Install dependencies
import importlib.util
import subprocess
import sys
from pathlib import Path

def run_python_module(*args):
    return subprocess.check_call([sys.executable, "-m", *args])

def ensure_pip_available():
    try:
        run_python_module("pip", "--version")
        return
    except subprocess.CalledProcessError:
        print("pip not found in this kernel environment. Bootstrapping with ensurepip ...")
    run_python_module("ensurepip", "--upgrade")
    run_python_module("pip", "install", "--upgrade", "pip")

required = {
    "litellm": "litellm==1.74.9",
    "pydantic": "pydantic==2.11.7",
    "openai": "openai==1.93.0",
}

missing = [name for name in required if importlib.util.find_spec(name) is None]
requirements_file = Path("requirements.txt")

if missing:
    print(f"Missing packages: {missing}")
    ensure_pip_available()
    if requirements_file.exists():
        print("Installing from requirements.txt ...")
        run_python_module("pip", "install", "-r", str(requirements_file))
    else:
        print("requirements.txt not found. Installing inline package list ...")
        run_python_module("pip", "install", *required.values())
else:
    print("All required packages already installed.")
```

```python
# Workshop Step 3: Validate dependencies and API key variables
import os
import importlib

modules = ["litellm", "pydantic", "openai"]
for m in modules:
    importlib.import_module(m)

provider = os.environ.get("LLM_PROVIDER", "openai").lower()
model = os.environ.get("LLM_MODEL", "gpt-4o-mini")
openai_key = os.environ.get("OPENAI_API_KEY", "")
ollama_base_url = os.environ.get("OLLAMA_BASE_URL", "http://localhost:11434/v1")

print("Import check: OK")
print(f"Provider={provider}, Model={model}")
print(f"OPENAI_API_KEY present: {bool(openai_key)}")
print(f"OLLAMA_BASE_URL: {ollama_base_url}")

if provider == "openai" and not openai_key:
    print("Warning: LLM_PROVIDER=openai but OPENAI_API_KEY is not set.")
if provider == "ollama":
    print("Ollama selected. Start Ollama locally before running the model cell.")
```
